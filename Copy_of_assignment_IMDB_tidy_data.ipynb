### Tidying the data

Next, you will further tidy the two dataframes **stars** and **genres**.

Let's start with **stars**.

```python
# Transform the dataframe "stars" to a new dataframe named "stars_long" with the following four columns:
# Series_Title, Released_Year, StarNo (e.g., Star1, Star2, ...), StarName
# Hint: use melt()
stars_long = stars.melt(id_vars=['Series_Title', 'Released_Year'], value_vars=['Star1', 'Star2', 'Star3', 'Star4'], var_name='StarNo', value_name='StarName')
stars_long
```

```python
# Can you transform dataframe 'stars_long' back to its original shape?
# Hint: use pivot()
stars_long_pivot = stars_long.pivot(index=['Series_Title', 'Released_Year'], columns = 'StarNo', values='StarName').reset_index()
stars_long_pivot.columns.name = None
stars_long_pivot
```

```python
# In dataframe "stars_long", find all movies that star "Morgan Freeman".
stars_long[stars_long['StarName'] == 'Morgan Freeman']['Series_Title']
```

```python
# Who appeared in Star2 the most times? List the top five actors.
stars_long[stars_long['StarNo'] == 'Star2']['StarName'].value_counts().head(5)
```

```python
# Who stars in the most movies in this list? List the top 20 actors.
stars_long['StarName'].value_counts().head(20)
```

```python
# Which movie stars had the highest total gross in this movie list? Show the top 10 actors.
# Hint: Join "stars_long" and "imdb"; then group by StarName
stars_long.merge(imdb, on=['Series_Title', 'Released_Year']).groupby('StarName')['Gross'].sum().head(10)
```

```python
# Find the best director-actor duos
# i.e., director-actor pairs that collaborated in at least five movies,
# sort them in descending order of average IMDB_Rating
# Hint: Join imdb and stars_long; group by ['Director','StarName']
best_duolist = imdb.merge(stars_long, on=['Series_Title', 'Released_Year']).groupby(['Director','StarName']).agg({'IMDB_Rating':'mean', 'Series_Title':'count'}).sort_values(by=['IMDB_Rating','Series_Title'], ascending=False).head(10)
best_duolist
```

```python
# Bonus question
# Who did "Amy Adams" co-star with in this movie list?
# Hint: Join stars_long with itself to find pairs of co-stars
amy_adams_movies = stars_long[stars_long['StarName'] == 'Amy Adams']
co_stars = amy_adams_movies.merge(stars_long, on=['Series_Title', 'Released_Year'], suffixes=('_amy', '_co'))

# Filter out Amy Adams herself and get unique co-star names
amy_adams_co_stars = co_stars[co_stars['StarName_co'] != 'Amy Adams']['StarName_co'].unique()
amy_adams_co_stars
```

Next, let's reshape the dataframe **genres**, which is a little bit more complicated.

```python
# Step 1: Split the 'Genre' string by ', ' into a list of individual genres and expand them to different columns
genres_split = genres['Genre'].str.split(', ', expand = True)
genres_split
```

```python
# Step 2: Rename the columns as: Genre1, Genre2, Genre3
genres_split.columns = ['Genre1', 'Genre2', 'Genre3']
genres_split
```

```python
# Step 3: Combine ['Series_Title','Released_Year'] in 'genres' and ['Genre1','Genre2','Genre3'] in 'genres_split'.
# Save it to a new dataframe named 'genres_wide'.
# Hint: Use pd.concat(...)
genres_wide = pd.concat([genres[['Series_Title', 'Released_Year']], genres_split], axis=1)
genres_wide
```

```python
# Step 4: Transform genres_wide to a new dataframe genres_long with the following four columns:
# Series_Title, Released_Year, GenreNo (e.g., Genre1, Genre2, Genre3), GenreName
# Hint: use melt()
genres_long = genres_wide.melt(id_vars=['Series_Title', 'Released_Year'], value_vars=['Genre1', 'Genre2', 'Genre3'], var_name='GenreNo', value_name='GenreName')
genres_long
```

```python
# How many unique genres (atomic values, e.g., Drama, Comedy, ...) are there?
# How many movies are there for each genre?
genres_long['GenreName'].value_counts() # there is 21 unique genres
```

```python
# What is the average IMDB rating for each genre?
# Sort the genres in descending order of average IMDB_Rating.
# Hint: join imdb with genres_long; group by GenreName
genres_long.merge(imdb, on=['Series_Title', 'Released_Year']).groupby('GenreName')['IMDB_Rating'].mean().sort_values(ascending=False)
```

```python
# Who is the "King of Comedy" (i.e., the actor who starred in the most comedy movies)?
# Hint: find all comedies in genres_long ; join with stars_long; group by StarName
genres_long[genres_long['GenreName'] == 'Comedy'].merge(stars_long, on=['Series_Title', 'Released_Year']).groupby('StarName').size().sort_values(ascending=False)
```

```python
# Who are the best action stars?
# i.e., the actors who star in at least 5 action movies, sorted by average IMDB rating in descending order
# Hint: You need to join columns from all three dataframes: imdb, stars_long, genres_long
genres_long[genres_long['GenreName'] == 'Action'].merge(stars_long, on=['Series_Title', 'Released_Year']).merge(imdb, on=['Series_Title', 'Released_Year']).groupby('StarName')['IMDB_Rating'].mean().sort_values(ascending=False)
```

```python
# Create a pivot table to show the number of movies of different genres in different decades.
# Row: GenreName
# Column: Decade
# Hint: https://pandas.pydata.org/docs/reference/api/pandas.pivot_table.html
pivotTable = pd.pivot_table(genres_long.merge(imdb[['Series_Title', 'Released_Year', 'Decade']], on=['Series_Title', 'Released_Year'], how='left'),
                           values='Series_Title',
                           index='GenreName',
                           columns='Decade',
                           aggfunc='count')
pivotTable
```
